As colunas desse dataset são:
- Country	[País]: 	 Nome do país.
- Density (P/Km2)	[Densidade]: 	 Densidade populacional medida em pessoas por quilômetro quadrado.
- Abbreviation	[Abreviatura]: 	 Abreviatura ou código que representa o país.
- Agricultural Land (%)	[Terra Agrícola]: 	 Percentagem da área de terra utilizada para fins agrícolas.
- Land Area (Km2)	[Área Terrestre]: 	 Área total do país em quilômetros quadrados.
- Armed Forces Size	[Tamanho das Forças Armadas]: 	 Tamanho das forças armadas do país.
- Birth Rate	[Taxa de natalidade]: 	 Número de nascimentos por 1.000 habitantes por ano.
- Calling Code	[Código de chamada]: 	 Código de chamada internacional para o país.
- Capital/Major City	[Capital/Cidade Principal]: 	 Nome da capital ou cidade principal.
- CO2 Emissions	[Emissões de CO2]: 	 Emissões de dióxido de carbono em toneladas.
- CPI	[IPC]:	 Índice de Preços ao Consumidor, uma medida de inflação e poder de compra.
- CPI Change (%)	[Variação do IPC]: 	 Variação percentual do Índice de Preços ao Consumidor em relação ao ano anterior.
- Currency_Code	[Códico_moeda]: 	 Código da moeda utilizada no país.
- Fertility Rate	[Taxa de fertilidade]: 	 Número médio de filhos nascidos de uma mulher durante sua vida.
- Forested Area (%)	[Área Florestada]: 	 Porcentagem de área coberta por florestas.
- Gasoline_Price	[Preco_Gasolina] 	 Preço da gasolina por litro em moeda local.
- GDP	[PIB]: 	 Produto Interno Bruto, valor total dos bens e serviços produzidos no país.
- Gross Primary Education Enrollment (%)	[Matrícula bruta no ensino primário]: 	 Taxa bruta de matrícula no ensino primário.
- Gross Tertiary Education Enrollment (%)	[Matrícula bruta no ensino superior]: 	 Taxa bruta de matrícula no ensino superior.
- Infant Mortality	[Mortalidade Infantil]: 	 Número de mortes por 1.000 nascidos vivos antes de completar um ano de idade.
- Largest City	[Maior Cidade]: 	 Nome da maior cidade do país.
- Life Expectancy	[Expectativa de vida]: 	 Número médio de anos que se espera que um recém-nascido viva.
- Maternal Mortality Ratio	[Razão de Mortalidade Materna]: 	 Número de mortes maternas por 100.000 nascidos vivos.
- Minimum Wage	[Salário Mínimo]: 	 Nível do salário mínimo em moeda local.
- Official Language	[Idioma Oficial]: 	 Língua(s) oficial(is) falada(s) no país.
- Out of Pocket Health Expenditure (%)	[Despesas Desembolsadas com Saúde]: 	 Percentagem das despesas totais com saúde pagas diretamente pelos indivíduos.
- Physicians per Thousand	[Médicos por Mil]: 	 Número de médicos por mil pessoas.
- Population	[População]: 	 População total do país.
- Population:  Labor Force Participation (%)	[População:	 Participação na Força de Trabalho (%)]: Porcentagem da população que faz parte da força de trabalho.
- Tax Revenue (%)	[Receita Fiscal]: 	 Receita tributária em percentagem do PIB.
- Total Tax Rate	[Taxa Total de Imposto]: 	 Carga tributária global como percentagem dos lucros comerciais.
- Unemployment Rate	[Taxa de Desemprego]: 	 Porcentagem da força de trabalho que está desempregada.
- Urban Population	[População Urbana]: 	 Porcentagem da população que vive em áreas urbanas.
- Latitude	[Latitude]: 	 Coordenada de latitude da localização do país.
- Longitude	[Longitude]: 	 Coordenada de longitude da localização do país.

```python
# Contando a quantidade de valores nulos de forma gráfica
fig, ax = plt.subplots()

ax.bar(base.isnull().sum().sort_values(ascending=False).index, base.isnull().sum().sort_values(ascending=False).values)

ax.tick_params(axis='x',labelrotation=90)

plt.show()
```
