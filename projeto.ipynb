## 2 - Preparando os Dados

```python
#ajuste das colunas numéricas que estão definidas como categóricas

transform_features = [ 'Agricultural Land( %)',  'Armed Forces size','Land Area(Km2)','Density(P/Km2)', 
'Forested Area (%)', 'Co2-Emissions','CPI', 'CPI Change (%)', 'Gasoline Price', 'GDP',
'Gross primary education enrollment (%)','Gross tertiary education enrollment (%)', 
'Minimum wage', 'Out of pocket health expenditure', 'Urban_population','Population: Labor force participation (%)',
'Population' ,'Tax revenue (%)', 'Unemployment rate','Total tax rate']

for column in transform_features:
    df[column]=df[column].astype(str)
    df[column]=df[column].str.replace(",","")
    df[column]=df[column].str.replace("$","")
    df[column]=df[column].str.replace("%","").astype(float)

import warnings     # Controle de avisos

warnings.filterwarnings('ignore') # Nunca exibe avisos correspondentes
```

```python
#novas features

df["GDP_per_capita"]=df["GDP"]/df["Population"]
df["Rural_population"]=df["Population"]-df["Urban_population"]
df["urban_pop/total_pop"] = df["Urban_population"]/df["Population"]
df["rural_pop/total_pop"] = df["Rural_population"]/df["Population"]
df["Agricultural Land Area (Km2)"] = df["Land Area(Km2)"]*df['Agricultural Land( %)']
```

```python
#unindo o dataset de continentes ao dataset de países

merged_df = pd.merge(df, continents_df, on='Abbreviation', how='left')
df = merged_df
df.T
```

### 4.1 - Visualização de dados com regressão simples

- Para mostrar rapidamente as relações entre os recursos, utilizaremos uma ferramenta de regressão simples da biblioteca Plotly. Com isso, selecionamos o método dos Mínimos Quadrados Ordinários (OLS) para fazer isso. Criaremos uma linha de tendência baseada em OLS e ela nos dará informações mais precisas sobre como essas características estão correlacionadas e quão bem a regressão descreve a relação entre elas. Resumindo, uma linha de tendência linear é criada para descrever o comportamento dos dados. Onde: Y=a∗X+b

- Onde Y
   é um recurso selecionado e X
   outro. As constantes a
   e b
   pode nos informar sobre quão intensa é a relação entre os recursos.

#### 4.1.1 - Indicadores de saúde

```python
# Relação entre médicos por mil, expectativa de vida e taxa de natalidade

plt = px.scatter(df, x="Physicians per thousand", y="Birth Rate", color = "Life expectancy", log_x="True", trendline="ols",
                 trendline_options=dict(log_x=True),color_continuous_scale="Spectral",hover_data=df[['Country']],
                 title="Taxa de Natalidade e Expectativa de Vida por Médicos por Mil")

plt.update_xaxes(title_text="Médicos por Mil (Mpm)")
plt.update_yaxes(title_text="Taxa de Natalidade")

plt.show()
```

- O valor de $R^2$ varia de 0 a 1, em que 1 afirma que o modelo descreve perfeitamente a relação entre as características. Os valores $R^2$ são calculados por:
$$
   R^2=1−\frac{Σ(y_i−y_{(tl)})^2}{Σ(y_i−y_{avg})^2}
$$

- Em que o $y_i$ são os pontos de dados y localização, $y_{tl}$ a linha de tendência e $y_{avg}$ o valor médio dos pontos.

- Por este gráfico, podemos ver que os países com elevada taxa de natalidade e menor MpM são também os países com menor esperança de vida. Vale a pena analisar esses dados relacionados a indicadores econômicos, como o PIB per capita que criamos. O $R^2$ é muito bom, mostrando que o modelo está bem ajustado. Além disso, a curva é descrescente e o coeficiente $a$ da regressão é negativo, o que significa que a relação entre a Taxa de Natalidade e o MpM é inversamente proporcional.


```python
plt = px.scatter(df, x="GDP_per_capita", y="Birth Rate", color = "Life expectancy",
                 trendline="ols", trendline_options=dict(log_x=True), color_continuous_scale="Spectral",hover_data=df[['Country']],
                 title="Taxa de Nascimento e Expectativa de Vida por PIB per capita")

plt.update_xaxes(title_text="PIB per capita")
plt.update_yaxes(title_text="Taxa de Natalidade")

plt.show()
```

- Como podemos ver neste gráfico, a expectativa de vida tende a aumentar em países com maior PIB per capita, o que indica populações ricas em média. Além disso, taxas de natalidade mais elevadas estão relacionadas com países com um menor PIB per capita e, como vimos nos gráficos anteriores, menos PpT. Portanto, o País de Baixa Renda pode não ter como investir em uma boa estrutura de saúde, o que pode resultar em Elevada Mortalidade Infantil e Materna. Aqui $R^2=0,69$ para Mortalidade Infantil e $R^2=0,63$ para Mortalidade Materna. O coeficiente $a$ ainda são negativos, indicando relação inversalmente proporcional.

```python
fig = make_subplots(rows=2, cols=1)


plt2= px.scatter(df, x="Physicians per thousand", y="Infant mortality", color = "Life expectancy",
                 trendline="ols", trendline_options=dict(log_x=True), color_continuous_scale="Spectral", hover_data=df[['Country']])


fig.add_trace(go.Scatter(plt2.data[0]),row=1,col=1)
fig.add_trace(go.Scatter(plt2.data[1],line=dict(color='#EF553B')),row=1,col=1)

fig.update_yaxes(title_text="Mortalidade Infantil", row=1, col=1)

plt3= px.scatter(df, x="Physicians per thousand", y="Maternal mortality ratio", color = "Life expectancy",
                 trendline="ols", trendline_options=dict(log_x=True), color_continuous_scale="Spectral", hover_data=df[['Country']])


fig.add_trace(go.Scatter(plt3.data[0]),row=2,col=1)
fig.add_trace(go.Scatter(plt3.data[1],line=dict(color='#EF553B')),row=2,col=1)

fig.update_xaxes(title_text="Médicos por Mil (Mpm)", row=2, col=1)
fig.update_yaxes(title_text="Mortalidade Materna", row=2, col=1)

fig.show()
```

- O mesmo ocorre pros gráficos abaixo, agora relacionando a Mortalidede Infantil e Materna com o PIB per capita. Aqui $R^2=0,62$ para Mortalidade Infantil e $R^2=0,47$ para Mortalidade Materna. O coeficiente $a$ ainda são negativos, indicando relação inversalmente proporcional.

```python
fig = make_subplots(rows=2, cols=1)


plt2= px.scatter(df, x="GDP_per_capita", y="Infant mortality", color = "Life expectancy",
                 trendline="ols", trendline_options=dict(log_x=True), color_continuous_scale="Spectral", hover_data=df[['Country']])


fig.add_trace(go.Scatter(plt2.data[0]),row=1,col=1)
fig.add_trace(go.Scatter(plt2.data[1],line=dict(color='#EF553B')),row=1,col=1)

fig.update_yaxes(title_text="Mortalidade Infantil", row=1, col=1)

plt3= px.scatter(df, x="GDP_per_capita", y="Maternal mortality ratio", color = "Life expectancy",
                 trendline="ols", trendline_options=dict(log_x=True), color_continuous_scale="Spectral", hover_data=df[['Country']])


fig.add_trace(go.Scatter(plt3.data[0]),row=2,col=1)
fig.add_trace(go.Scatter(plt3.data[1],line=dict(color='#EF553B')),row=2,col=1)

fig.update_xaxes(title_text="PIB per capita", row=2, col=1)
fig.update_yaxes(title_text="Taxa de mortalidade materna", row=2, col=1)

fig.show()
```
